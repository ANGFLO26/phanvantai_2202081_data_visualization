Lượng Nhập Cư vào Canada (1980 - July 2024)

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import warnings
warnings.simplefilter('ignore', FutureWarning)
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
file_path = "/home/phanvantai/Documents/THREE_YEARS/three_year_02/data visualization/canada/canadian_immigration_data1980-2024.csv"
df = pd.read_csv(file_path)
df
```

```python
df_filtered = df[df["Country"] != "Total"]

# Calculate the total number of immigrants to Canada per year (from 1980 to 2024)
years = np.arange(1980, 2025)
total_immigration_per_year = df_filtered[years.astype(str)].sum()

# Update the number of immigrants for 2024 by adding data from "July - 2024"
total_immigration_per_year["2024"] += df_filtered["July - 2024"].sum()

# Identify the year with the highest immigration
max_year = total_immigration_per_year.idxmax()
max_value = total_immigration_per_year.max()

# Extract valid year data for plotting
total_immigration_values = total_immigration_per_year.loc[years.astype(str)].values

# Plot the graph
plt.figure(figsize=(12, 6))
plt.plot(years, total_immigration_values, marker="o", linestyle="-", color="b", label="Number of Immigrants")
plt.axvline(x=int(max_year), color="r", linestyle="--", alpha=0.7, label=f"Highest: {max_year} ({max_value:,})")
plt.scatter(int(max_year), max_value, color="red", zorder=3)

# Set title and labels
plt.title("Immigration to Canada (1980 - July 2024)")
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show the plot
plt.show()

# Print the year with the highest immigration
print(f"The year with the highest immigration: {max_year}, with {int(max_value):,} immigrants.")
```

```python
# Remove the 'Total' row as it does not represent a specific country
df_filtered = df[df["Country"] != "Total"]

# Calculate the total number of immigrants to Canada per year (from 1980 to 2024)
years = np.arange(1980, 2025)
total_immigration_per_year = df_filtered[years.astype(str)].sum()

# Update the number of immigrants for 2024 by adding data from "July - 2024"
total_immigration_per_year["2024"] += df_filtered["July - 2024"].sum()

# Identify the year with the highest immigration
max_year = total_immigration_per_year.idxmax()
max_value = total_immigration_per_year.max()

# Extract valid year data for plotting
total_immigration_values = total_immigration_per_year.loc[years.astype(str)].values

# Set seaborn style
sns.set_style("whitegrid")
sns.set_palette("Blues_r")

# Plot the graph
plt.figure(figsize=(12, 6))
sns.lineplot(x=years, y=total_immigration_values, marker="o", label="Number of Immigrants")
sns.scatterplot(x=[int(max_year)], y=[max_value], color="red", s=100, label=f"Highest: {max_year} ({max_value:,})")
plt.axvline(x=int(max_year), color="r", linestyle="--", alpha=0.7)

# Set title and labels
plt.title("Immigration to Canada (1980 - July 2024)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Immigrants", fontsize=12)
plt.legend()
plt.show()

# Print the year with the highest immigration
print(f"The year with the highest immigration: {max_year}, with {int(max_value):,} immigrants.")
```
