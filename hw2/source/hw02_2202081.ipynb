```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('/content/drive/MyDrive/data visua/train.csv')
```

```python
train_data.head()
```

**mối quan hệ giữa SalePrice và GrLivArea (diện tích sàn sử dụng).**



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=train_data, x='GrLivArea', y='SalePrice')

# Thiết lập tiêu đề và nhãn cho trục
plt.title('Mối quan hệ giữa Diện tích sàn và Giá nhà', fontsize=14)
plt.xlabel('Diện tích sàn (GrLivArea)', fontsize=12)
plt.ylabel('Giá nhà (SalePrice)', fontsize=12)

# Hiển thị biểu đồ
plt.show()
```

**phân phối của SalePrice theo các nhóm OverallQual (chất lượng tổng thể)**

```python
# Vẽ biểu đồ hộp (boxplot) giữa 'OverallQual' và 'SalePrice'
plt.figure(figsize=(10, 6))
sns.boxplot(data=train_data, x='OverallQual', y='SalePrice')

# Thiết lập tiêu đề và nhãn cho trục
plt.title('Phân phối Giá nhà theo Chất lượng Tổng thể', fontsize=14)
plt.xlabel('Chất lượng Tổng thể (OverallQual)', fontsize=12)
plt.ylabel('Giá nhà (SalePrice)', fontsize=12)

# Hiển thị biểu đồ
plt.show()
```
