```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
import folium
from shapely.geometry import Point, Polygon
import plotly.express as px
from IPython.display import display
from folium.plugins import MarkerCluster
```

```python
file_path ="/home/phanvantai/Documents/THREE_YEARS/three_year_02/data visualization/document/Police_Department_Incidents_-_Previous_Year__2016_.csv"
```

```python
tables = pd.read_csv(file_path)
tables
```

```python
crime_by_district = tables["PdDistrict"].value_counts().reset_index()
crime_by_district.columns = ["PdDistrict", "Total Incidents"]
print(crime_by_district)
```

```python
tables_park = tables[(tables["PdDistrict"] == "PARK") & (tables["DayOfWeek"] == "Tuesday")]
tables_park
```

```python
print(f"Số lượng dữ liệu sau khi lọc: {len(tables_park)}")
print(tables_park[["Category", "Descript", "X", "Y"]].head()) 
```

```python
# Kiểm tra dữ liệu hợp lệ
if not tables_park.empty:
    # Loại bỏ các điểm có tọa độ X, Y không hợp lệ (X=0, Y=0)
    tables_park = tables_park[(tables_park["X"] != 0) & (tables_park["Y"] != 0)]
    
    # Kiểm tra lại nếu sau khi lọc dữ liệu vẫn còn
    if not tables_park.empty:
        # Lấy tọa độ trung tâm
        map_center = [tables_park["Y"].mean(), tables_park["X"].mean()]

        # Khởi tạo bản đồ
        crime_map = folium.Map(location=map_center, zoom_start=13)

        # Tạo Marker Cluster
        marker_cluster = MarkerCluster().add_to(crime_map)

        # Thêm từng marker vào cluster
        for _, row in tables_park.iterrows():
            folium.Marker(
                location=[row["Y"], row["X"]],
                popup=f"{row['Category']}: {row['Descript']}",
                icon=folium.Icon(color="red", icon="info-sign")
            ).add_to(marker_cluster)

        # Hiển thị bản đồ
        display(crime_map)  # Dùng display() nếu chạy trong Jupyter Notebook
    else:
        print("Dữ liệu sau khi lọc bị trống! Không có tọa độ hợp lệ để hiển thị.")
else:
    print("Không có dữ liệu phạm tội hợp lệ để hiển thị trên bản đồ.")
```

```python
# Chuyển cột 'Time' thành kiểu datetime nếu chưa có
tables_park['Time'] = pd.to_datetime(tables_park['Time'], format='%H:%M')
print("Tổng số bản ghi ban đầu:", len(tables_park))

# Cập nhật hàm phân loại khoảng thời gian
def categorize_time(hour):
    if 6 <= hour < 12:
        return "Buổi sáng (06:00 - 11:59)"
    elif 12 <= hour < 14:
        return "Buổi trưa (12:00 - 13:59)"
    elif 14 <= hour < 18:
        return "Buổi chiều (14:00 - 17:59)"
    elif 18 <= hour < 22:
        return "Buổi tối (18:00 - 21:59)"
    else:
        return "Buổi đêm (22:00 - 05:59)"  # Bao gồm cả đêm đến sáng hôm sau

# Thêm cột mới xác định khoảng thời gian
tables_park['TimePeriod'] = tables_park['Time'].dt.hour.apply(categorize_time)

# Đếm số lượng tội phạm theo khoảng thời gian
crime_by_time = tables_park['TimePeriod'].value_counts().reset_index()
crime_by_time.columns = ['Khoảng thời gian', 'Số vụ phạm tội']

# Sắp xếp theo đúng thứ tự thời gian
time_order = [
    "Buổi sáng (06:00 - 11:59)",
    "Buổi trưa (12:00 - 13:59)",
    "Buổi chiều (14:00 - 17:59)",
    "Buổi tối (18:00 - 21:59)",
    "Buổi đêm (22:00 - 05:59)"
]
crime_by_time['Khoảng thời gian'] = pd.Categorical(crime_by_time['Khoảng thời gian'], categories=time_order, ordered=True)
crime_by_time = crime_by_time.sort_values('Khoảng thời gian')

# Hiển thị bảng
crime_by_time
```

```python
tables_lunch = tables_park[tables_park["TimePeriod"] == "Buổi trưa (12:00 - 13:59)"]

# Loại bỏ tọa độ không hợp lệ
tables_lunch = tables_lunch[(tables_lunch["X"] != 0) & (tables_lunch["Y"] != 0)]

if not tables_lunch.empty:
    # Xác định trung tâm bản đồ
    map_center = [tables_lunch["Y"].mean(), tables_lunch["X"].mean()]
    
    # Khởi tạo bản đồ
    crime_map = folium.Map(location=map_center, zoom_start=13)
    
    # Tạo Marker Cluster
    marker_cluster = MarkerCluster().add_to(crime_map)

    # Thêm marker vào bản đồ
    for _, row in tables_lunch.iterrows():
        folium.Marker(
            location=[row["Y"], row["X"]],
            popup=f"{row['Category']}: {row['Descript']}",
            icon=folium.Icon(color="red", icon="info-sign")
        ).add_to(marker_cluster)

    # Hiển thị bản đồ
    display(crime_map)
else:
    print("Không có vụ phạm tội nào vào buổi trưa tại khu vực PARK vào thứ Ba.")
```
